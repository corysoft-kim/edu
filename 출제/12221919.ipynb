```python
### Retriever를 위한 라이브러리 불러오기
import warnings
warnings.filterwarnings('ignore')
from transformers import AutoTokenizer, AutoModel
import torch
```

### Step A-1 Retriever Class 구현

🔴[주의사항] Retriever 구현시 sentence_transfomers 등 외부 Library 활용 금지

🔴[주의사항] Retriever 구현시 이미 정의된 변수명(self.documents, self.documents_embeddings 등) 수정 금지

Q A-1) get_embedding(text) 구현

 - 함수 인자로 받은 text에 대한 **tensor** 타입의 embedding vector인 `torch.Size=([self.model.config.hidden_size])`반환

Q A-2) build_embeddings_from_doc_titles() 구현

 - self.documents에 대한 **tensor** 타입의 embedding matrix 반환
   1. get_embedding()를 사용하여, 각 document(제품 spec) 마다 **tensor** 타입의 embedding vector 생성

      🔴[주의사항] document 별 embedding vector는 **제품명(첫번째 line)**만 사용하여 생성할 것
   2. document들의 embedding vector를 통합하여 `torch.Size=([len(self.documents), self.model.config.hidden_size])`의 **tensor** 반환

Q A-3) calculate_euclidean_distances(query_embedding) 구현

 - 함수 인자로 받은 query_embedding과 self.documents_embeddings 간의 euclidean distance 값들을 갖는 `torch.Size=([len(self.documents)])`의 **tensor** 반환

   🔴[주의사항] tensor의 기본 연산만 사용하여야 함
   
   🔴[주의사항] torch.linalg, torch.cdist, torch.dist, torch.norm, torch.nn.functional.pairwise_distance, scipy, sklearn 등 쉽게 계산해주는 Library 사용 금지

Q A-4) calculate_dot_products(query_embedding) 구현

 - 함수 인자로 받은 query embedding과 self.documents_embeddings 간의 dot_product(내적) 값들을 갖는 `torch.Size=([len(self.documents)])`의 **tensor** 반환

   🔴[주의사항] 2중 반복문 (예: for, while) 로직 사용 금지

   🔴[주의사항] torch Library 내 tensor 연산만 사용하고, scipy, sklearn, tensorflow 등 외부 Library 사용 금지

Q A-5) retrieve_doc(query) 구현

 - 함수 인자로 받은 query와 self.documents_embeddings 기준으로 가장 유사한 document 1개 반환

   🔴[주의사항] 유사도 척도로는 Q A-3과 Q A-4에서 구현한 calculate_euclidean_distances() 와 calculate_dot_products() 중에 **하나를 선택**하여 사용하되, 모든 query에 대해서 정확한 검색 결과를 제공하는 것을 선택해야함
   
   🔴[주의사항] euclidean는 "distance"이므로 두 vector가 유사할수록 값이 작아짐에 유의


```python
import torch
from torch import Tensor

class Retriever:
    def __init__(self, documents, embedding_model_name):
        self.documents = documents
        self.tokenizer = AutoTokenizer.from_pretrained(embedding_model_name)
        self.model = AutoModel.from_pretrained(embedding_model_name)
        self.documents_embeddings = self.build_embeddings_from_doc_titles()

    def get_embedding(self, text) -> Tensor:
        ### Q A-1) 작성 필요
        tokenized_text = self.tokenizer(text, return_tensors="pt")
        squeezed_last_hidden_state = self.model(**tokenized_text).last_hidden_state.squeeze()

        embedding = squeezed_last_hidden_state.mean(dim=0)
        return embedding

    def build_embeddings_from_doc_titles(self) -> Tensor:
        ### Q A-2) 작성 필요
        documents_embeddings = torch.zeros([len(self.documents), self.model.config.hidden_size])
        for i, doc in enumerate(self.documents):
          documents_embeddings[i] = self.get_embedding(doc.split('\n')[0])
        return documents_embeddings

    def calculate_euclidean_distances(self, query_embedding) -> Tensor:
        ### Q A-3) 작성 필요
        euclidean_distances=(torch.sum((query_embedding - self.documents_embeddings)**2, dim=1)).sqrt()
        return euclidean_distances

    def calculate_dot_products(self, query_embedding) -> Tensor:
        ### Q A-4) 작성 필요
        dot_products=torch.sum(query_embedding * self.documents_embeddings, dim=1)
        return dot_products

    # def retrieve_doc(self, query) -> str:
    #     query_embedding = self.get_embedding(query)
    #     scores = self.calculate_dot_products(query_embedding)
    #     top_1_idx = torch.argmax(scores)

    #     return self.documents[top_1_idx] # top_1_idx : 함수 인자로 받은 query와 가장 유사한 document의 index

    def retrieve_doc(self, query) -> str:
        query_embedding = self.get_embedding(query)
        scores = self.calculate_euclidean_distances(query_embedding)
        top_1_idx = torch.argmin(scores)

        return self.documents[top_1_idx] # top_1_idx : 함수 인자로 받은 query와 가장 유사한 document의 index
```

# Step B. Finetune

```python
query = queries[0] ### 하나의 제품에 대한 질문만 수행
context = contexts[0]

### Retriever 결과 확인 ###
pretty_print("QUESTION", query)
pretty_print("CONTEXT",context)

### sLM을 위한 prompt 정의###
prompt = build_prompt(context=context, query=query)
pretty_print("PROMPT",prompt)
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, TrainingArguments, Trainer, DataCollatorForLanguageModeling
from datasets import DatasetDict

path_pretrained_model = "./models/pretrained_model/"
path_finetuned_model = "./models/finetuned_model/"
path_dataset = "./data/training_data.tsv"

device_map = "cpu"
# hyperparams for generation
max_new_tokens = 80
```

```python
# Tokenizing and generating a response
def generate_response(prompt, model, tokenizer):
    tokenized_prompt = tokenizer([prompt], return_tensors='pt')
    generated_ids = model.generate(tokenized_prompt.input_ids,
                   tokenizer=tokenizer,
                   max_new_tokens=max_new_tokens)
    #Stripping the input text from generated_ids
    generated_ids = [
        output_ids[len(input_ids):] for input_ids, output_ids in zip(tokenized_prompt.input_ids, generated_ids)
    ]

    response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
    return response
```

### Step B-1 현재 Pretrained 모델의 학습전 Response 확인

```python
# Loading the pretrained model
model = AutoModelForCausalLM.from_pretrained(
    path_pretrained_model,
    device_map=device_map
)

#Loading the tokenizer
tokenizer = AutoTokenizer.from_pretrained(
    path_pretrained_model
)
response = generate_response(prompt, model, tokenizer)
print("response(before training):", response)
```

Q B-1) training dataset을 로드하여 DatasetDict 객체를 생성하세요 (split: "train")

Q B-2) dataset(DatasetDict) 내 Dataset의 features로 input_ids, token_type_ids, attention_mask를 가질 수 있도록, lambda 함수 내에 알맞은 값을 입력하세요

```python
# Loading the dataset
from datasets import load_dataset
dataset = load_dataset('./data/')

### Q B-2) 작성 필요
dataset = dataset.map(lambda x: tokenizer(x["text_data"], truncation=True, padding=True, max_length=512), batched=True, remove_columns=dataset["train"].column_names)
```

Q B-3) DataCollatorForLanguageModeling 객체를 생성하세요 (적절한 params 사용)

```python
# Constructing a collator
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False, return_tensors='pt')
```

Q B-4) 제시된 값들을 이용하여 TrainingArguments 객체를 생성하세요
- learning rate: 3e-4
- epoch: 30
- batch size: 2
- cpu 사용: True
- report_to: "none"
- logging_steps : 40

```python
# Setting TrainingArguments
training_args = TrainingArguments(
    learning_rate=3e-4,
    num_train_epochs=30,
    per_device_train_batch_size=2,
    use_cpu=True,
    report_to='none',
    logging_steps=40,
    output_dir='out'
)
```

Constructing Trainer and Training the model

```python
# Constructing trainer, training and saving the model.
trainer = Trainer(model, train_dataset=dataset['train'], args=training_args, data_collator=data_collator)
trainer.train()

trainer.save_model(path_finetuned_model)
```

### Step C-1 api call로 이미지 생성 수행

```python
def call_api_dalle(target =""):
    import urllib.request
    from openai import OpenAI
    import matplotlib.pyplot as plt
    import matplotlib.image as img
    client = OpenAI(api_key=open_api_key)
    response = client.images.generate(
        prompt=f'{target} 광고 이미지를 생성해줘. 제품과, 포스터 배경에는 {target}을 강조하는 배경을 채워줘',
        model='dall-e-3',
        quality='hd',
        size='1024x1792',
        n=1
    )
    image_url = response.data[0].url
    image_path = f"./{str(target)}_result.png"
    urllib.request.urlretrieve(image_url,image_path)
    print("Image_Saved",image_path)

    plt.imshow(img.imread(image_path))
    plt.show()

    return image_path
```
